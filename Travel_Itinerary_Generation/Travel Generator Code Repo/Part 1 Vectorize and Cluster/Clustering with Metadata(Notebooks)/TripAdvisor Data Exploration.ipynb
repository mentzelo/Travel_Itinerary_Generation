```python
restaurants_df = df[df['type'] == 'RESTAURANT']
attractions_df = df[df['type'] == 'ATTRACTION']
hotels_df = df[df['type'] == 'HOTEL']
```

First let's clean the restaurants table

```python


##Below function removes all columns with NaN over threshold, and removes columns with # of duplicates threshold

def clean_dataframe(df, threshold_nan=0.99, threshold_dup=.99):
    filtered_df = df.loc[:, (df.isnull().mean() < threshold_nan)]
    removed_dup_cols = []
    for col in filtered_df.columns:
        if filtered_df[col].value_counts().max() / filtered_df.shape[0] > threshold_dup:
            removed_dup_cols.append(col)

    cleaned_df = filtered_df.drop(columns=removed_dup_cols)
    return cleaned_df


##This function takes all columns that start with a particular string, and concatenates them together in a 
##new column with a comma and space
def merge_columns(df, new_name, start = '', end = ''):
    all_columns = [col for col in df.columns if (col.startswith(start) and col.endswith(end))]
    df[new_name] = df[all_columns].apply(lambda x: ', '.join(x.astype(str)), axis=1)
    df.drop(columns=all_columns, inplace=True)
    return df

restaurants_df = merge_columns(restaurants_df, 'owner_image_url', start='', end = 'image_url')
restaurants_df = merge_columns(restaurants_df, 'features', 'features')
restaurants_df = merge_columns(restaurants_df, 'review_Tags', 'reviewTags', end = 'text')
restaurants_df = merge_columns(restaurants_df, 'cuisines', 'cuisines')
restaurants_df = merge_columns(restaurants_df, 'meal_Types', 'mealTypes')
restaurants_df = merge_columns(restaurants_df, 'featuredReviewtext', 'ownersTopReasons/topReasons', end = 'text')
restaurants_df = merge_columns(restaurants_df, 'dishes', 'dishes')
```

```python
#Clean the dataframe: removes columns if there isn't enough data.
#restaurants_df =  clean_dataframe(restaurants_df, threshold_nan=0.99, threshold_dup=.99)
```

```python
##We now have the restaurant_df and we need to get rid of irrelevant columns like nearest station for now:
## We want to keep:webUrl	website	features	review_Tags	cuisines	meal_Types	dishes
## address, description, email, name, menuWebUrl, website,  numberOfReviews, priceLevel, priceRange, rankingString
## ranking denominator, rating, rawRanking
restaurants_df = restaurants_df[['webUrl', 'website', 'features', 'review_Tags', 'cuisines', 'meal_Types', 'dishes', 'address', 
                                'description', 'email', 'name', 'menuWebUrl',  'numberOfReviews', 'priceLevel', 'priceRange', 'rankingString',
                                'rankingDenominator', 'rating', 'rawRanking', 'image', 'owner_image_url']]


```

```python
#Now let's get rid of restaurants that are rated below 3.5
#This gives us around 1000 restaurants in the Athens area, should be good enough to work with.
restaurants_df = restaurants_df.loc[restaurants_df['rawRanking'] > 3.5]

restaurants_df
```

## Hotels

Now we will take a look at the hotels and do a bit of data cleaning

```python
hotels_df.head()
```

```python
hotels_df = merge_columns(hotels_df, 'review_Tags', 'reviewTags', end = 'text')
hotels_df = merge_columns(hotels_df, 'amenities', 'amenities')
hotels_df = merge_columns(hotels_df, 'roomTips', 'roomTips', end = 'text')


hotels_df = hotels_df[['webUrl', 'website', 'amenities','roomTips', 'review_Tags', 'address', 
                                'description', 'email', 'name',  'numberOfReviews', 'priceLevel', 'priceRange', 'rankingString',
                                'rankingDenominator', 'rating', 'rawRanking', 'image']]
```

```python
## Remove hotels with ranking below 3.5
hotels_df = hotels_df.loc[hotels_df['rawRanking'] > 3.5]
```

## Attractions

Now we will take a look at the attractions

```python
attractions_df = merge_columns(attractions_df, 'tour_Offers', 'offerGroup', end = 'title')
attractions_df = merge_columns(attractions_df, 'review_Tags', 'reviewTags', end = 'text')
```

```python
attractions_df = attractions_df[['webUrl', 'website', 'review_Tags', 'address', 
                                'description', 'name',  'numberOfReviews', 'tour_Offers', 'rankingString',
                                'rankingDenominator', 'rating', 'rawRanking','subtype/0', 'image']]
```

```python
##Remove attractions that are low rated
attractions_df = attractions_df.loc[attractions_df['rawRanking'] > 3.5]
```
